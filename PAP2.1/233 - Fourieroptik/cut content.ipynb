# Fit der Beugungsfunktion an die Messdaten

### Einzelspalt: Hauptmaximum

```python
#Import Beugungsbild
p_s_b_0, i_s_b_0 = np.loadtxt('einzelspalt beugung hauptmaximum.txt', skiprows=3, unpack=True)

mm_s_b_0 = p_s_b_0 * eich #px in mm

plt.plot(mm_s_b_0, i_s_b_0, color='red', linewidth=1)
plt.grid(alpha=0.5, linestyle='--')
plt.xlabel('x [mm]')
plt.ylabel('Intensit√§t / '+' b.E.')
plt.title('Abb. 1: Fraunhofer Beugung am Spalt')
#plt.savefig("FraunhoferBeug1.eps", format="eps")
```

```python
def fitFunc(x, a, scale, off, ug):
    return a*(np.sin(scale*(x-off))/(scale*(x-off)))**2 +ug

init_vals1 = [1000, 15, 1.1, 10]
fitParams1, fitCovariances1 = curve_fit(fitFunc, mm_s_b_0, i_s_b_0, p0=init_vals1, maxfev=8000)
print(fitParams1)

plt.plot(mm_s_b_0, fitFunc(mm_s_b_0, *fitParams1), label= "Theorie")
plt.grid(alpha=0.5, linestyle='--')
plt.plot(mm_s_b_0[::8], i_s_b_0[::8], color='red', linestyle='None', marker='.', label= "Messung")
plt.xlabel('x [mm]')
plt.ylabel('Intensitaet [b.E.]')
plt.title('Fraunhofer Beugung')
plt.legend()
#plt.savefig("FrauenhoferBeugFit.eps", format="eps")
```

```python
plt.plot(mm_s_b_0, (fitFunc(mm_s_b_0, *fitParams1)-i_s_b_0), label= "Residuen: Theorie - Messung")
plt.xlabel('x [mm]')
plt.ylabel('Residuen')
plt.title('Fraunhofer Beugung')
plt.legend()
#plt.savefig("FrauenhoferBeugResid.eps", format="eps")
```
